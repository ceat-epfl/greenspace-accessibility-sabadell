# Spatial extent

```python
import contextily as cx
import osmnx as ox
```

```python
# ACHTUNG: for some reason, the `Sabadell` query returns a different boundary (i.e.,
# Badia del Vall√®s), so we use the osmid instead
# nominatim_query = "Sabadell"
nominatim_query = "R344202"
buffer_dist = 1000
dst_filepath = "../data/raw/spatial-extent.gpkg"
```

```python
# query, project and add buffer
place_gdf = ox.project_gdf(ox.geocode_to_gdf(nominatim_query, by_osmid=True)).buffer(
    buffer_dist
)
ax = place_gdf.plot(alpha=0.5)
cx.add_basemap(ax=ax, crs=place_gdf.crs)
```

```python
# dump to file
place_gdf.to_file(dst_filepath)
```
